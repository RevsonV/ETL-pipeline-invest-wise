```python
import pandas as pd
import json
from pandas_gbq import to_gbq
```

```python
# Creates a dataframe from the json file
with open('./json_files/BALANCE_SHEET.json') as json_file:
  data = json.load(json_file)
  data = data['quarterlyReports']
  df = pd.DataFrame(data)
```

```python
# Clean and prepare data to import to BigQuery
for col in df.columns:
  if col == 'fiscalDateEnding' or col == 'reportedCurrency':
    pass
  else:
    df[col] = df[col].apply(lambda x: x.replace('None', '0')).astype(int)
```

```python
project = 'investwise-417720'
table_id = 'company_data.balance_sheet'

# Imports dataframe to a BigQuery table
to_gbq(
    df,
    project_id=project,
    destination_table=table_id
    if_exists='replace'
)
```
